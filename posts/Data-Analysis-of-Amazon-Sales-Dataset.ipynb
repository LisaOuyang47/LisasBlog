```python
df.drop([282,324],inplace = True)
```

```python
# Convert the data type of prices, rating, rating_count, and discount percentage to int/float
```

```python
# replace unnecessary characters
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', '')
df['discounted_price'] = df['discounted_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('‚Çπ', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
```

```python
df.head()
```

```python
df['discounted_price'] = pd.to_numeric(df['discounted_price'])
df['actual_price'] = pd.to_numeric(df['actual_price'])
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'])
```

```python
df[df['rating'] == '|']
```

```python
df.drop(index = 1279, inplace= True)
```

```python
df['rating'] = pd.to_numeric(df['rating'])
df['rating_count'] = pd.to_numeric(df['rating_count'])
```

## Exploratory Data Visualizations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# What's the most common rating in this dataset?
df['rating'].plot(kind = 'hist', bins = 10, color = 'lightblue', title = 'Rating Frequency');
```

```python
# Boxplot of discount percentage
df.plot(y = 'discount_percentage', kind = 'box');
```

```python
# Percentage frequency of discounts for products with ratings greater than 4.0 
truncted_data = df.loc[df['rating'] > 4.0]

truncted_data.plot.hist(y = 'discount_percentage', color = 'lightblue', bins = 5);
```

```python
df['category'].unique()
```

```python
# Cleaning Product Category
df['grouped_category'] = df['category'].str.split('|').str[0]
df.drop('category', axis=1, inplace=True)
```

```python
df['grouped_category'].unique()
```

```python
grouped_data = df.groupby(['grouped_category'])
grouped_data1 = df.groupby(['grouped_category'])['grouped_category'].count()

grouped_data1.plot(y = 'grouped_category', kind = 'hist', 
                   bins = 10, color = 'bisque', 
                   title = 'Most frequent product category');

# need to polish-change x-axis' name?
```

```python
grouped_data1
```

```python
# Average product discount percentage per category
grouped_data['discount_percentage','grouped_category'].mean().sort_values(
    by='discount_percentage', ascending=False).plot(
    kind = 'bar', color = 'lightgreen');
```
